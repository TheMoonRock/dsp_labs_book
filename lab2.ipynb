# Преобразование Фурье
Прежде, чем перейти к анализу сигналов, надо разобраться, что такое преобразование Фурье. 
> Преобразование Фурье (символ ℱ) — операция, сопоставляющая одной функции вещественной переменной другую функцию вещественной переменной. Эта новая функция описывает коэффициенты («амплитуды») при разложении исходной функции на элементарные составляющие — гармонические колебания с разными частотами (подобно тому, как музыкальный аккорд может быть выражен в виде суммы музыкальных звуков, которые его составляют). [Wiki](https://ru.wikipedia.org/wiki/%D0%9F%D1%80%D0%B5%D0%BE%D0%B1%D1%80%D0%B0%D0%B7%D0%BE%D0%B2%D0%B0%D0%BD%D0%B8%D0%B5_%D0%A4%D1%83%D1%80%D1%8C%D0%B5)



Формула прямого дискретного преобразования Фурье (ДПФ, Discrete Fourier Transform, DFT):
$$ X_k = \sum_{n=0}^{N-1} x_{n} \cdot e^{-i \frac{2\pi}{N} k n} $$


# Задание 1 (2 балла)
Реализовать дискретное преобразование Фурье. Сравнить с реализацией *scipy.fft.fft*

```python
def DFT(x):
    # YOUR CODE HERE
    N = len(x)
    X = np.zeros(N, dtype=complex)  # Initialize the output array with complex numbers
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-2j * np.pi * k * n / N)  # Compute each frequency component
    return X

    # raise NotImplementedError()
```

Класическое ДПФ имеет существенный недостаток - сложность алгоритма $O(N^2)$. На данный момент существуют чуть более сложные реализации преобразования Фурье, которые избавляются от квадратичной сложности. Например, алгоритм быстрого преобразования Фурье (Fast Fourier Transform, FFT) имеет сложность $O(N log(N))$, что делает его значительно более быстрым на длинных последовательностях.

**Подсказка:** учтите, что для БПФ нам нужно, чтобы количество отсчетов сигнала было степенью двойки

```python
def FFT(x):
    # YOUR CODE HERE
    N = len(x)
    
    # Если длина сигнала не является степенью двойки, дополните его нулями
    if N & (N - 1) != 0:  # Проверка на степень двойки
        next_power_of_2 = 2 ** np.ceil(np.log2(N)).astype(int)  # Следующая степень двойки
        x = np.pad(x, (0, next_power_of_2 - N), 'constant')
        N = len(x)

    if N <= 1:
        return x

    # Разделение на четные и нечетные элементы
    even = FFT(x[0::2])
    odd = FFT(x[1::2])
    
    # Объединение результатов
    T = np.exp(-2j * np.pi * np.arange(len(odd)) / N) * odd
    
    # Объединяем результаты
    return np.concatenate([even + T, even - T])
    #raise NotImplementedError()
# Пример сигналов (вы можете определить свои)
signals = {
    "signal1": np.random.rand(1024),  # Случайный сигнал длиной 1024
    "signal2": np.sin(2 * np.pi * np.arange(1024) / 64),  # Синусоида длиной 1024
}

# Запуск тестов для всех сигналов
for name, sig in signals.items():
    print(f"Проверка '{name}'")
    assert test_DTF(sig, transform=FFT) == 0, "Проверьте вашу реализацию"
print("Все в порядке!")
```

### 1.Почему амплитуда всех гармоник единичного импульса равна константе?

# YOUR ANSWER HERE
Амплитуда всех гармоник единичного импульса равна константе, потому что единичный импульс (или дельта-функция) содержит все частоты с одинаковой амплитудой. Это связано с тем, что в частотной области дельта-функция представляется как плоский спектр, где каждая гармоника имеет одинаковую величину. Таким образом, преобразование Фурье единичного импульса дает постоянную амплитуду для всех частот.

### 2. Какие выводы можно сделать смотря на амплитудный спектр? А на фазовый?

# YOUR ANSWER HERE
- Амплитудный спектр: Он показывает, какие частоты присутствуют в сигнале и с какой силой. Если в спектре есть пики, это означает, что соответствующие частоты доминируют в сигнале. Например, если вы видите резкий пик на определенной частоте, это указывает на то, что эта частота является основной компонентой сигнала.
- Фазовый спектр: Он показывает фазовые сдвиги каждой гармоники относительно других. Фаза важна для понимания временных отношений между компонентами сигнала. Если фаза изменяется резко, это может указывать на резкие изменения в сигнале.

### 3. Как перевести номер отсчета в частоту?

# YOUR ANSWER HERE
Чтобы перевести номер отсчета в частоту, можно использовать следующую формулу:

$$
f = \frac{n \cdot f_s}{N}
$$

где:
- $f$ — частота в герцах (Hz),
- $n$ — номер отсчета (индекс в спектре),
- $f_s$ — частота дискретизации (количество выборок в секунду),
- $N$ — общее количество выборок.

Эта формула позволяет определить соответствующую частоту для каждого индекса в амплитудном или фазовом спектре.

### 4. Что в фазовом спектре является полезной информацией, а что - шумом? Почему?

# YOUR ANSWER HERE
В фазовом спектре полезной информацией являются значения фазы для основных гармоник сигнала. Эти значения помогают понять временные отношения между компонентами сигнала и их взаимодействие.

Шумом могут быть значения фазы для высокочастотных компонентов или для компонентов с низкой амплитудой. Эти значения могут не иметь физического смысла и могут быть результатом случайных колебаний или ошибок измерения. Обычно такие фазы не влияют на общий анализ сигнала и могут быть проигнорированы.

# Оконное преобразование Фурье
Преобразование Фурье делает переход из временной области в частотную, позволяя анализировать составляющие сигнал гармоники. Однако это не всегда удобно. В реальных сигналах зачастую составляющие их гармоники изменяются со временем. Например, каждая буква в слове имеет свой набор основных гармоник, и простое преобразование Фурье не сильно поможет нам определить, что было сказано. Для того, чтобы анализировать такие сигналы, обычно используется [Оконное преобразование Фурье](https://ru.wikipedia.org/wiki/Оконное_преобразование_Фурье). Такое преобразование позволяет анализировать гармоники сигнала в коротких промежутках времени и сохранить временную ось при построении спектрограммы.
# Задание 3 (2 балла)
Реализовать оконное преобразование Фурье (Short-time Fourier transform), пользуясь только библиотеками *numpy* и *scipy*. Сравнить с *librosa.stft*


```python
def STFT(x, n_fft=2048, hop_length=512, window='hann'):
    # x - signal
    # n_fft - fft window size 
    # hop_length - step size between ffts
    # window - window type. See scipy.signal.get_window
    # return spectrogram 
    
    # YOUR CODE HERE
    # Получаем окно
    # Получаем окно
    if isinstance(window, str):
        window = scipy.signal.get_window(window, n_fft)
    
    # Центрируем сигнал
    x = np.pad(x, (n_fft // 2, n_fft // 2), mode='constant')
    
    # Количество окон
    n_windows = (len(x) - n_fft) // hop_length + 1
    
    # Инициализация спектрограммы
    Zxx = np.zeros((n_fft // 2 + 1, n_windows), dtype=complex)
    
    for i in range(n_windows):
        start = i * hop_length
        end = start + n_fft
        
        # Получаем текущее окно с учетом границ
        x_windowed = x[start:end] * window
        
        # Выполняем FFT и сохраняем результат
        Zxx[:, i] = np.fft.rfft(x_windowed)
    
    return Zxx

    # raise NotImplementedError()
    
```

```python
def test_stft(x, n_fft=2048, hop_length=512, window='hann'):
    librosa_result = librosa.stft(x, n_fft=n_fft, hop_length=hop_length, window='hann', center=True)
    result = STFT(x, n_fft=n_fft, hop_length=hop_length, window='hann')
    if librosa_result.shape != result.shape:
        print(f"Your shape {result.shape} != librosa stft shape {librosa_result.shape} ")
        return -1 
    if not np.allclose(librosa_result, result):
        print(f"Wrong results. Diff {np.abs(librosa_result-result)}")
        return -2
    return 0
        
for name, sig in signals.items():
    print(f"Checking '{name}'")
    for n_fft in (8, 4):
        for hop_length in (n_fft//4, n_fft//2):
            assert test_stft(sig, n_fft=n_fft, hop_length=hop_length)==0, \
                f"Test failed. Params n_fft {n_fft}. Hop_length {hop_length}"
    print("ok")
print("All ok!")
```
